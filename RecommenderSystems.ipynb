```python
#Recommender Systems
#Content Filtering(personal recommendation) and Collaborative Filtering(recommendation based on other users)
#Collaborative Filtering
#We use previous behaviors, similar users, other users
# movies- IMDB
import pandas
movie_names = pandas.read_csv("https://modcom.co.ke/datasets/Movie_Id_Titles.csv")
movie_names.head(5)
```

```python
movie_ratings = pandas.read_csv("https://modcom.co.ke/datasets/file.csv")
movie_ratings.head(50)
```

```python
movie_ratings = pandas.read_csv("https://modcom.co.ke/datasets/file.csv")
movie_ratings
```

```python
#merge the two
data = pandas.merge(movie_ratings,movie_names, on = 'item_id')
data
```

```python
# Find the average rating per movie
data.groupby('title')['rating'].mean().sort_values(ascending=False)
```

```python
#How many people rated each movie
data.groupby('title')['rating'].count().sort_values(ascending=False)
```

```python
#combine the average rating with the number of ratings
mean_ratings = pandas.DataFrame(data.groupby('title')['rating'].mean())
mean_ratings['num_of_ratings'] = data.groupby('title')['rating'].count()
mean_ratings
```

```python
# Pivot Tabe = analyzes large amounts of dasta in detail
pivot = data.pivot_table(index = 'user_id', columns = 'title', values = 'rating')
pivot
```

```python
# Select a movie I want to get recommendation
selected = pivot['Toy Story (1995)']
similar = pivot.corrwith(selected)
similar
# create a dataframe
similar_df = pandas.DataFrame(similar, columns = ['Correlations'])
similar_df.sort_values('Correlations', ascending = False)

#We will consider a correlation of above 0.5 and number of ratings >200
```

```python
#Combine above dataframe  with mean_ratings
similar_df = similar_df.join(mean_ratings['num_of_ratings'])
similar_df
```

```python
#finally get all that have over 200 number of ratings
similar_df[similar_df['num_of_ratings'] >200].sort_values('Correlations',
                                                          ascending = False).head(20)
```
