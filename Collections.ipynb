```python
import pandas as pd
data =pd.read_excel("/content/drive/MyDrive/Datasets/September 2021 Collections.xlsx")
data
```

```python
#check empties
data.isnull().sum()
```

```python
data['Portfolio Manager'].fillna('Rita', inplace= True)
data.isnull().sum()
```

```python
data['Agent'].fillna('Office', inplace= True)
data.isnull().sum()
```

```python
#Graphs with seaborn
import seaborn as sns
import matplotlib.pyplot as plt
print(plt.style.available)
```

```python
plt.style.use('ggplot')
sns.stripplot(x = 'Portfolio Manager', y= 'Monthly Collections', data= data)
plt.title("Portfolio Manager vs Monthly Collections in KES")
```

```python
plt.style.use('ggplot')
sns.barplot(x = 'Portfolio Manager', y= 'Monthly Collections', data= data)
plt.title("Portfolio Manager vs Monthly Collections in KES")
```

```python
plt.style.use('ggplot')
sns.barplot(x = 'Portfolio Manager', y= 'Monthly Collections',estimator = sum,  palette = 'Blues', data = data)
plt.title("Portfolio Manager vs Monthly Collections in KES")
```

```python
#lineplot
sns.lineplot(x = "Loan amount", y = "Monthly Collections",  data = data)
plt.title("Loan Amount by Monthly Collections")
```

```python
plt.style.use('seaborn')
sns.lmplot(x = 'Loan amount', y= 'Monthly Collections', hue = 'Status', palette= "magma", data= data)
plt.title("Loan Amount vs Monthly Collections in KES")
```

```python
plt.style.use('seaborn')
sns.scatterplot(x = 'Monthly Collections', y= 'Expected collections', hue = 'Status', palette= "flare", data= data)
plt.title("Monthly Collections vs Expected Collections in KES")
```

```python
sns.displot(x ="Loan amount",color ="blue" , data = data)
```

```python
sns.distplot(data["Loan amount"], color= "red")
```

```python
from scipy.stats import  norm
sns.distplot(data["Loan amount"], color= "red", fit= norm, kde= True)
plt.title(" Loan Amount Distribution")
plt.ylabel("Freq")
```

```python
sns.histplot(data = data, x="Loan amount", color= "red", hue ="Status")
```
