```python
import pandas
data = pandas.read_csv("http://modcom.co.ke/datasets/bank_full.csv")
data.head(10)
```

```python
#ML RULES
#Machine Learning doesnt work with empties- Remove empties/ fill them - imputation
# Machine Learning does not work with texts- Convert texts to digits- Encoding
# Check Empties
data.isnull().sum()
```

```python
#Fill Emptoies
data['job'].fillna('unknown', inplace=True)
data['marital'].fillna('unknown', inplace=True)
data['loan'].fillna('unknown', inplace=True)
#for duration its a continous variable  we find the mean
duration_mean = data['duration'].mean()
data['duration'].fillna(duration_mean, inplace=True)

#for dropping records use dropna
data.isnull().sum()
```

```python
#loan to Digits
data['loan'].replace({'no':0, 'yes':1, 'unknown':2}, inplace=True)
data['housing'].replace({'no':0, 'yes':1, 'unknown':2}, inplace=True)
data['default'].replace({'no':0, 'yes':1, 'unknown':2}, inplace=True)
data['contact'].replace({'cellular':0, 'telephone':1,}, inplace=True)
#divorced,married,single,unknown
data['marital'].replace({'divorced':0, 'married':1, 'single':2, 'unknown':3}, inplace=True)
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
data['job']= encoder.fit_transform(data['job'])
data['education']= encoder.fit_transform(data['education'])
data['month']= encoder.fit_transform(data['month'])
data['day+AF8-of+AF8-week']= encoder.fit_transform(data['day+AF8-of+AF8-week'])
data['poutcome'].replace({'failure':0,'nonexistent':1, 'success':2}, inplace=True)
data.head(50)
```

```python
#lambda function
def slice(param):
  if '-' in param:
   answer = param.split("-")
   return answer[1]
  else:
    return param

data['cons.conf.idx'] = data["cons.conf.idx"].apply(slice)
data['emp.var.rate'] = data["emp.var.rate"].apply(slice)
data.groupby("cons.conf.idx").size()
```

```python
array = data.values
X = array[:,0:20]#other features
Y = array[:,20 ] # 
```

```python
from sklearn import model_selection
X_train, X_test, Y_train,Y_test = model_selection.train_test_split(X,Y, test_size = 0.3,
                                                                   random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.svm import SVC

#Cross Validation
# It is a resampling technique that uses different portions of your data to test and train a model using different iterations
#Iteration
#Type1: KFOLD - Used to estimate the skills of the model with the current data we have.

models = []
models.append(("DT", DecisionTreeClassifier()))
models.append(("LR", LogisticRegression(solver = "liblinear")))
models.append(("RF", RandomForestClassifier ()))
models.append(("GB", GradientBoostingClassifier()))
models.append(("KNN", KNeighborsClassifier()))
models.append(("Gaussian", GaussianNB()))
models.append(("DT", DecisionTreeClassifier()))
 
for name, model in models:
   kfold = model_selection.KFold(n_splits=10, random_state=42, shuffle=True)
   cv_results = model_selection.cross_val_score(model,X_train,Y_train, cv =kfold, scoring='accuracy')
   print(name, cv_results.mean())
```

```python
model = GradientBoostingClassifier()
model.fit(X_train, Y_train)

GradientBoostingClassifier()
```

```python
#predict xtest
predictions = model.predict(X_test)
print("Predicted", predictions)
print("Expected", Y_test)
```

```python
#metrics
from sklearn.metrics import accuracy_score
print("Accuracy score", accuracy_score(predictions, Y_test))
```

```python
#test
from sklearn.metrics import classification_report
print(classification_report(predictions,Y_test))
```
