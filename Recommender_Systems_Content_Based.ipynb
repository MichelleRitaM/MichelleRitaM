```python
import pandas
data = pandas.read_csv('https://modcom.co.ke/datasets/shared_articles.csv')
data
```

```python
# We will only work with the text column
X = data['text'].values
#X
```

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('distilbert-base-nli-stsb-mean-tokens')

embeddings = model.encode(X, show_progress_bar =True)
embeddings
```

```python
df = pandas.DataFrame(embeddings)
df
#machinelearningplus.com/nlp/cosine-similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim_data = cosine_similarity(df)

#pass above to be like a dataframe
cosine_data = pandas.DataFrame(cosine_sim_data)
cosine_data
```

```python
#find similarity
def recommendation(newsid):
  index_recommendations = cosine_data.loc[newsid].sort_values(ascending=False).index.tolist()[1:10]
  print("Similar Indexes", index_recommendations)
  news_titles = data['title'].loc[index_recommendations].values
  print("Similar Articles", news_titles)

recommendation(100)
```
